```python
import requests
import os

# Pexels API 키 설정
PEXELS_API_KEY = 'TgHgjrXFlKQQuR7dmtRU0R7Z3kcHG7v9R6usQRU1ZtBFqyoCZNFerGEr'

# 이미지를 다운로드할 폴더 생성
download_folder = './crawled_img/pexels'
os.makedirs(download_folder, exist_ok=True)

# Pexels API에서 이미지 다운로드
def download_pexels_images(query, total_count=80, per_page=80):
    num_pages = total_count // per_page
    remainder = total_count % per_page

    for page in range(num_pages + (1 if remainder > 0 else 0)):
        url = f'https://api.pexels.com/v1/search?query={query}&per_page={per_page}&page={page + 1}'
        headers = {'Authorization': PEXELS_API_KEY}
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            data = response.json()
            for photo in data['photos']:
                img_url = photo['src']['original']
                img_data = requests.get(img_url).content
                img_name = os.path.join(download_folder, f"{photo['id']}.jpg")

                with open(img_name, 'wb') as img_file:
                    img_file.write(img_data)

                print(f"Downloaded: {img_name}")

    print("Download completed.")

# 이미지 다운로드 호출
download_pexels_images(query='galleries', total_count=80, per_page=80)
```
