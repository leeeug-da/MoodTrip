```python
import requests
import json
import os

API_KEY = '4DkGnOep6O2Y1yMG5BxLo3FfU2e1PKlf9ZLzXZzA9OU'  # Unsplash에서 발급받은 API 키
query = "quiet"  # 검색할 키워드
maxImages = 1500  # 이미지 숫자 ex) 1500
path = "./crawled_img/unsplash/" + query  # 저장할 폴더

if not os.path.exists(path):
    os.makedirs(path)

success = 0
page = 1
per_page = 30  # Unsplash API의 경우 한 번에 최대 30개의 이미지를 로드함

while success < maxImages:
    url = f'https://api.unsplash.com/search/photos?query={query}&page={page}&per_page={per_page}&client_id={API_KEY}'
    res = requests.get(url)
    data = json.loads(res.text)

    for img in data['results']:
        img_url = img['urls']['small']  # 원하는 이미지의 크기를 선택
        filename = f'{query}_{success + 1}.jpg'
        saveUrl = os.path.join(path, filename)

        with open(saveUrl, 'wb') as f:
            f.write(requests.get(img_url).content)
        success += 1

        if success >= maxImages:
            break

    page += 1

print('다운로드 성공 : ' + str(success))
```
