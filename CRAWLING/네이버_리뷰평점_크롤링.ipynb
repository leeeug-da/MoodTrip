```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import numpy as np
import os
from tqdm import tqdm_notebook
```

```python
driver = webdriver.Chrome() 

# 엑셀 파일 읽기
df = pd.read_excel('./관광지정보_대분류데이터/관광지.xlsx')

# '명칭' 컬럼의 값을 이용하여 검색어 생성
df['naver_keyword'] = df['명칭'].str.replace(' ', '%20')
df['naver_map_url'] = ''

for i, keyword in enumerate(df['naver_keyword'].tolist()):
    print("이번에 찾을 키워드 :", i, f"/ {df.shape[0] -1} 행", keyword)
    try:
        naver_map_search_url = f"https://m.map.naver.com/search2/search.naver?query={keyword}&sm=hty&style=v5"
        driver.get(naver_map_search_url)
        time.sleep(3.5)
        
        try:
            for j in range(1, 100):
                text = driver.find_element_by_css_selector(f'#ct > div.search_listview._content._ctList > ul > li:nth-child({str(j)}) > div.item_info > div.item_info_inn > div').text # 2
                if text[5:] in df['주소'][i]: # 1
                    df.iloc[i,-1] = driver.find_element_by_css_selector(f'#ct > div.search_listview._content._ctList > ul > li:nth-child({str(j)}) > div.item_info > a.a_item.a_item_distance._linkSiteview').get_attribute('data-cid') # 2
                    break
                else:
                    continue
            
        except NoSuchElementException:
            print("해당 요소를 찾을 수 없습니다.")
            df.iloc[i,-1] = np.nan
            continue
        
    except Exception as e1:
        print("다음 에러가 발생했습니다: ", e1)
        df.iloc[i,-1] = np.nan
        continue

driver.quit()

df['naver_map_url'] = "https://m.place.naver.com/place/" + df['naver_map_url']
df = df.loc[~df['naver_map_url'].isnull()]  # null값이 아닌 행만 선택
```

```python
# 데이터를 미리 리스트에 담은 다음, 마지막에 데이터 프레임에 합치기
naver_map_type_list = []
review_qty_list = []
naver_map_star_review_stars_list = []

# 드라이버: 별점, 리뷰 개수 크롤링
driver = webdriver.Chrome()

# 기존 코드는 이 부분부터 시작
for i, url in enumerate(tqdm_notebook(df['naver_map_url'])):

    driver.get(url)
    try:
        naver_map_type = driver.find_element_by_css_selector("#_title > div > span.Fc1rA").text  # 관광지 이름
        naver_map_type_list.append(naver_map_type)
    except Exception as e1:
        naver_map_type_list.append('NaN')
    
    try:
        review_qty = driver.find_element_by_css_selector("#app-root > div > div > div > div.place_section.no_margin.OP4V8 > div.zD5Nm > div.dAsGb").text  # 모든 리뷰 수 => 후처리 필요
        review_qty_list.append(review_qty)
    except Exception as e1:
        review_qty_list.append(0)
        
    try:        
        star_review_stars = driver.find_element_by_css_selector("#app-root > div > div > div > div.place_section.no_margin.OP4V8 > div.zD5Nm > div.dAsGb > span.PXMot.LXIwF").text  # 별점
        naver_map_star_review_stars_list.append(star_review_stars)
    except Exception as e1:
        naver_map_star_review_stars_list.append(0)

driver.quit()
```

```python
# 새로운 데이터프레임 생성
new_df = pd.DataFrame()
new_df['name'] = df['명칭']  # .xlsx에 명시된 이름
new_df['naver_place_name'] = naver_map_type_list  # 관광지 이름
new_df['naver_star_point'] = naver_map_star_review_stars_list  # 별점
new_df['naver_review_qty'] = review_qty_list  # 모든 리뷰 수

# df 확인
new_df.head()
```

```python
# 후처리
import re

# '별점\n' 부분 제거
new_df['naver_star_point'] = new_df['naver_star_point'].apply(lambda x: x if x == 0 else x.replace('별점\n', ''))

# 'naver_review_qty' 열에서 '방문자리뷰' 또는 '블로그리뷰' 숫자를 추출하여 더한 값을 'naver_review_qty' 열에 넣기
def extract_and_sum_reviews(row):
    # 정수를 예상하는 부분에서는 정수가 아니면 0으로 설정
    visit_review = re.search(r'방문자리뷰 (\d+)', str(row['naver_review_qty']))
    blog_review = re.search(r'블로그리뷰 (\d+)', str(row['naver_review_qty']))

    visit_qty = int(visit_review.group(1)) if visit_review else 0
    blog_qty = int(blog_review.group(1)) if blog_review else 0
    
    return visit_qty + blog_qty

# apply 함수 사용 시, axis=1을 사용하여 각 행에 함수를 적용
new_df['naver_review_qty'] = new_df.apply(extract_and_sum_reviews, axis=1)
```

```python
# 인덱스 reset
new_df = new_df.reset_index()
# 최종 new_df 확인
new_df
```

```python
# csv로 저장
new_df.to_csv('./crawled_data/관광지.csv')
```
